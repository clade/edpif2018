* Lean by examples : http://matplotlib.org/gallery.html

```python
%matplotlib inline
# Or %pylab
# Or import matplotlib.pyplot as plt; plt.ion()
from matplotlib.pyplot import *
from numpy import linspace, sin, cos
```

## Functions : 
* plot, loglog, semilogx, semilogy
* xlabel, ylabel, title
* Latex like
* label and legend
* xlim, ylim
* figure
* subplot
* imshow
* savefig : svg, pdf

```python
t = linspace(0,10,1000)
figure('name', figsize=(10,10))
subplot(2,1,1)
plot(t, sin(t), label="sin")
plot(t, 0.5*cos(t), '--', label='cos')
xlabel('time (s)')
ylabel('$V_\epsilon$ (V)')
legend(loc='best')
ylim(-1.1, 1.1)
grid()
text(4, 0.5, '$E = mc^2$')
subplot(2,2,3)
plot(t, sin(t)**2)
tight_layout()
savefig('/tmp/my_figure.pdf')
savefig('/tmp/my_figure.svg')
```
