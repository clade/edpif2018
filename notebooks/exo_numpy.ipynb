```python
import numpy
def mandel(c):
    z = zeros(c.shape, dtype='complex')
    output = numpy.zeros(c.shape, dtype='int')
    for i in range(100):
        z[output==0] = z[output==0]**2 + c[output==0]
        output[(abs(z)>2) & (output==0)] = i
    return output

imshow(mandel(c))
```
