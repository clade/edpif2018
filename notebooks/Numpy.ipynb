```python
def my_function(omega, omega_0, width, amplitude):
    return amplitude/(1+(omega-omega_0)**2/width**2)

x = linspace(-10, 10, 1000)
y = my_function(x, 1, 3, 2.1)
plot(x, y)
```

```python
from numpy import vectorize

def my_function(x):
    if x<0:
        return -x
    if x<1:
        return x
    return 2-x

my_vect_function = vectorize(my_function)

x_axis = linspace(-5, 5, 101)
#y = [my_function(x) for x in x_axis]
y = my_vect_function(x_axis)
plot(x_axis, y)
```

```python
from numpy import vectorize

@vectorize
def my_function(x):
    if x<0:
        return -x
    if x<1:
        return x
    return 2-x

my_function(1)
```
